# 1.2 Correlation Analysis

### Q1.2.1 Compute the correlation matrix of all features.

```python
import pandas as pd
df = pd.read_csv("/Users/theodorsjetnanutvik/Desktop/MaskinLÃ¦ring/assignment1/data/WineQT.csv")

# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Show correlations with quality
corr_matrix["quality"].sort_values(ascending=False)
```

---

### 1.2.2 Plot a heatmap of the correlation matrix.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix (Wine Dataset)")
plt.savefig("../results/corr_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

### 1.2.3 Which variable has the strongest positive correlation with quality? Which variable has the strongest negative correlation with quality?

Reading from the correlation matrix we can find that:
- The strongest positive correlation with **quality** is **alcohol**
    $0.484866$
- The strongest negative correlation with **quality** is **volatile acidity**
    $-0.407394$
