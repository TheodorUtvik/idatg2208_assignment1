### Q 1.1.2 Which features show the highest variation based on summary statistics?

```python
# Calculate coefficient of variation (std / mean) for each feature
variation = (df.std() / df.mean()).sort_values(ascending=False)

print("Coefficient of Variation (sorted):")
print(variation)

# Optional: bar plot
import matplotlib.pyplot as plt

variation.drop("Id", errors="ignore").plot(kind="bar", figsize=(10,5))
plt.ylabel("Coefficient of Variation (std / mean)")
plt.title("Relative Variation of Features")
plt.tight_layout()
plt.show()
```

Reviewing the numbers after finding the **Coefficient of Variation**, the features that show the highest variations are:

- item Citric acid
- item Total sulfur dioxide
- item Free sulfur dioxide
